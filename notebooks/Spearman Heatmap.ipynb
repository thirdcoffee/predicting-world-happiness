```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

# Read data from Excel file
data = pd.read_excel('/workspaces/predicting-world-happiness/data/WHI/Complete.xlsx', sheet_name='Complete')

# List of variables
variables = [
    'Happiness Score', 'GDP per capita', 'Social Support', 'Life Expectancy',
    'Human Rights', 'CPI', 'World Giving', 'Rural Population', 'Consumer Inflation',
    'Military Expenditure', 'Population Growth', 'CO2 Emissions', 'Energy Use'
]

# Convert data to numeric and handle missing values
data[variables] = data[variables].apply(pd.to_numeric, errors='coerce')

# Calculate correlation matrix
correlation_matrix = data[variables].corr(method='spearman')

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)

# Set plot title and axis labels
plt.title('Correlation Matrix')
plt.xlabel('Variables')
plt.ylabel('Variables')

# Show the heatmap
plt.show()
```

```python
import pandas as pd
from scipy.stats import spearmanr

# Read data from Excel file
data = pd.read_excel('/workspaces/predicting-world-happiness/data/WHI/Complete.xlsx', sheet_name='Complete')


# List of variables
variables = [
    'Happiness Score', 'GDP per capita', 'Social Support', 'Life Expectancy',
    'Human Rights', 'CPI', 'World Giving', 'Rural Population', 'Consumer Inflation',
    'Military Expenditure', 'Population Growth', 'CO2 Emissions', 'Energy Use'
]

# Calculate correlation matrix
correlation_matrix = pd.DataFrame(index=variables, columns=variables)

for var1 in variables:
    for var2 in variables:
        values1 = data[var1].dropna()
        values2 = data[var2].dropna()
        common_indices = values1.index.intersection(values2.index)
        
        correlation, _ = spearmanr(values1[common_indices], values2[common_indices])
        correlation_matrix.loc[var1, var2] = correlation

# Display the correlation matrix
correlation_matrix
```
